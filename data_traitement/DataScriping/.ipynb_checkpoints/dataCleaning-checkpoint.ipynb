```python
import pandas
import csv
import numpy as np
import ast
import collections
import matplotlib.pyplot as plt
```

```python
colnames = ['ingredients']
data = pandas.read_csv('test.csv', usecols=colnames)
```

```python
table = []

for k,v in data.items():
    for x in v:
        try:   
            if x!=np.nan:
                res = ast.literal_eval(x)
                table.append(res)
        except:
            print(x)
    #print(res['nom_ingre'])
print(table)
```

```python

#print(res)
list_ingre = []
for n in table:
    #print(n)
    for x in n:
        list_ingre.append(x['nom_ingre'])
counter=collections.Counter(list_ingre)
#print(counter) 
```

```python
# Data to plot
labels = []
sizes = []
for x, y in counter.items():
    if y >=100:
        labels.append(x)
        sizes.append(y)

# Plot
plt.pie(sizes, labels=labels, autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)

plt.axis('equal')
plt.show()
```

```python
# Data to plot
labels = []
sizes = []

for x, y in counter.items():
    if y <100 and y >=50:
        labels.append(x)
        sizes.append(y)

# Plot
plt.pie(sizes, labels=labels)

plt.axis('equal')
plt.show()
```
